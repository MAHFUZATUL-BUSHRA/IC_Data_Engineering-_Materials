# TASK 5: Functions

```python
# 5a. Average function
def calculate_average(numbers):
    return sum(numbers) / len(numbers)

print(calculate_average([80, 92, 75, 88, 95]))
```

```python
# 5b. Grade function

def get_grade(score=0):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

print(get_grade(92))
print(get_grade(75))
print(get_grade(45))
```

```python
# 5c. Summary function

def summarize(scores):
    count = len(scores)
    avg = sum(scores) / count
    return count,avg

count, avg = summarize([85, 90, 78, 92])
print(f"Count: {count}, Average: {avg}")
```

# TASK 6: File Handling

```python
import csv, json
```

Write CSV file

```python
students = [
    {"name": "Rahim", "dept": "CSE", "cgpa": 3.65},
    {"name": "Fatima", "dept": "EEE", "cgpa": 3.82},
    {"name": "Karim", "dept": "CSE", "cgpa": 3.21},
    {"name": "Nadia", "dept": "BBA", "cgpa": 3.90},
    {"name": "Arif", "dept": "EEE", "cgpa": 2.95},
]
```

```python
with open("student.csv", "w", newline="") as file:
  student=csv.DictWriter(file, fieldnames=["name","dept","cgpa"])
  student.writeheader()
  student.writerows(students)
```

### Read CSV file

```python
with open("student.csv","r") as file:
  reading= csv.DictReader(file)
  for row in reading:
    print(row)
```

### Summary in Json

```python
cgpa = [student["cgpa"] for student in students]
```

```python
summary={
      "total_students": len(students),
      "average_cgpa": round(calculate_average(cgpa),2),
      "highest_cgpa": max(cgpa),
      "lowest_cgpa": min(cgpa),
}
```

### Write json

```python
with open("summary.json", "w") as file:
  json.dump(summary, file, indent=4)
```

```python
# Verify JSON
with open("summary.json") as j:
  result=json.load(j)
  print(result)
```
