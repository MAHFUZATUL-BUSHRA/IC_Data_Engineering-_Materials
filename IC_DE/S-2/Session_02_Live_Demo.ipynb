## 2. List Methods: Add, Remove, Sort

```python
# Adding items
sources = ["api", "csv"]
print("echo 1:", sources)
sources.append("database")        # add to end
print("echo 2:",sources)

sources.append("database2")        # add to end
print("echo 3:", sources)

sources.insert(1, "webhook")      # insert at position
print("echo 4:", sources)

sources.insert(6, "webhook2")      # insert at position
print("echo 5:", sources)
```

```python
# Removing items
sources.remove("csv")             # remove by value
print(sources)

last = sources.pop()              # remove & return last
print(f"Popped: {last}")
print(sources)
```

```python
# Sorting & searching
scores = [85, 92, 78, 95, 88]

scores.sort()
print(f"Sorted: {scores}")

scores.sort(reverse=True)
print(f"Descending: {scores}")

print(f"Min: {min(scores)}")
print(f"Max: {max(scores)}")
print(f"Sum: {sum(scores)}")
print(f"'api' in sources: {'api' in sources}")
print(f" 92 in scores: { 92 in scores}")
print(f" 91 in scores: { 91 in scores}")
```

## 3. List Comprehensions

```python
tables = ["users", "orders", "products"]

# Traditional loop — 4 lines
upper_tables = []
for m in tables:
    # print(m) #DEBUG
    # print(m.upper()) #DEBUG
    upper_tables.append(m.upper())
    # print(upper_tables) #DEBUG
print(f"Loop: {upper_tables}")

# List comprehension — 1 line! Same result
upper_tables_ls = [m.upper() for m in tables]
print(f"Comprehension: {upper_tables_ls}")
```

```python
row_counts = [5000, 12000, 3500, 800, 25000]
big_tables = []
for c in row_counts:
    if c > 4000:
        big_tables.append(c)
print(f"Loop - Big tables: {big_tables}")

# With condition (filter)
row_counts = [5000, 12000, 3500, 800, 25000]
big_tables = [c for c in row_counts if c > 4000]
print(f"List Comprehension - Big tables: {big_tables}")
```

## Sets and Tuples: Unique & Immutable Collections

```python
# Creating tuples
coords_list = [23.8103, 90.4125]   # Dhaka
coords_tuple = (23.8103, 90.4125)   # Dhaka
rgb = (255, 128, 0)

print(coords_list)
print(coords_tuple)

# Access like lists
print(coords_tuple[0])    # 23.8103
print(coords_list[0]) 

# But you CAN'T modify them
coords_list[0] = 100 
print(coords_list)
# coords_tuple[0] = 100   # TypeError!

# Unpacking (very common!)
lat, lng = coords_tuple
print(f"Lat: {lat}, Lng: {lng}")
```

```python
# Creating sets — duplicates removed automatically
sources = {"api", "csv", "api", "db"}
print(sources)   # {'api', 'csv', 'db'}

# Remove duplicates from a list
ids = [1, 2, 3, 2, 1, 4]
unique = list(set(ids))
print(f"Unique: {unique}")

# Set operations
a = {1, 2, 3}
b = {2, 3, 4}
print(f"Intersection: {a & b}")   # {2, 3}
print(f"Union: {a | b}")          # {1, 2, 3, 4}
print(f"Difference: {a - b}")     # {1}
print(f"Difference: {b - a}")     # {4}
```

## 4. Dictionaries: Key-Value Pairs

```python
# Creating a dictionary
pipeline = {
    "name": "daily_sales_etl",
    "status": "running",
    "rows_processed": 15000,
    "is_active": True
}

print(pipeline)
```

```python
# Accessing values
print(pipeline["name"])             # "daily_sales_etl"
print(pipeline.get("status"))       # "running"
print(pipeline.get("owner", "unknown"))  # "unknown" (default)
```

```python
# Why .get() is safer than []
# pipeline["owner"]   # KeyError! crashes your program
print(pipeline.get("owner", "not set"))  # safe, returns default
```

## 6. Functions: Reusable Code Blocks

```python
# Define a function
def calculate_success_rate(total, failed):
    """Calculate pipeline success rate."""
    success = total - failed
    rate = (success / total) * 100
    return round(rate, 2)

# Call the function
result = calculate_success_rate(10000, 150)
print(f"Success rate: {result}%")
```

```python
# Reuse it with different data!
print(calculate_success_rate(5000, 23))
print(calculate_success_rate(100000, 4500))
print(calculate_success_rate(800, 0))
```

## 9. Working with CSV Files

```python
import csv
# Writing CSV from data
results = [
    {"name": "Ali", "score": 95},
    {"name": "Sara", "score": 88},
    {"name": "Rafi", "score": 92},
]

with open("results.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["name", "score"])
    writer.writeheader()
    writer.writerows(results)

print("results.csv written!")
```

## 10. Working with JSON Files

```python
import json

# Writing JSON
config = {
    "pipeline": "sales_etl",
    "source": "crm_database",
    "tables": ["users", "orders", "products"],
    "settings": {
        "batch_size": 1000,
        "retry": 3
    }
}

with open("config.json", "w") as f:
   json.dump(config, f, indent=2) # dict → file

print("config.json written!")
```

```python
# Reading JSON
with open("config.json") as f:
    config = json.load(f) # file → dict

print(config, type(config))
print(f"Pipeline: {config['pipeline']}")
print(f"Tables: {config['tables']}")
print(f"Batch size: {config['settings']['batch_size']}")
```

```python
# JSON from/to string (common with APIs)
json_string = '{"status": "ok", "count": 42}'
data = json.loads(json_string)     # string -> dict
print(f"The dict is {data}", type(data))
print(data["count"])

back_to_string = json.dumps(data)  # dict -> string
print(back_to_string, type(back_to_string))
```

## 11. Putting It All Together: Mini ETL Pipeline

```python
import csv, json

def extract(filepath):
    """Read CSV and return list of dicts."""
    with open(filepath) as f:
        return list(csv.DictReader(f))

def transform(records):
    """Clean and enrich data."""
    cleaned = []
    for r in records:
        r["score"] = int(r["score"])
        r["name"] = r["name"].strip().title()
        if r["score"] > 0:
            cleaned.append(r)
    return cleaned

def load(data, output_path):
    """Save as JSON."""
    with open(output_path, "w") as f:
        json.dump(data, f, indent=2)

# Run the pipeline!
raw = extract("results.csv")
print(f"Extracted: {len(raw)} records")

clean = transform(raw)
print(f"Transformed: {len(clean)} records")

load(clean, "clean_results.json")
print(f"Loaded to clean_results.json!")
```

```python
# Verify the output
with open("clean_results.json") as f:
    result = json.load(f)

for item in result:
    print(f"{item['name']}: {item['score']}")
```
